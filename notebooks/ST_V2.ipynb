```python
import json
import pandas as pd
import numpy as np

from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')
from scipy import spatial



```

```python
#FUNCTIONS
def load_data(file_name):
  with open(file_name) as json_file:
      dataset = json.load(json_file)
      skill_discs = list(dataset.values())
      skills = list(dataset.keys())
  return skills,skill_discs

def create_df(skills,skill_discs):
    #Preparing the dataframe
    sim_df = pd.DataFrame(columns=skills)
    key_dict = {"key":skills}
    key_df = pd.DataFrame(key_dict)
    data_df = pd.concat([key_df,sim_df],axis=1)
    data_df = data_df.set_index('key')

    # import itertools
    # lst = list(itertools.repeat(0, len(skills)))
    # # print(lst)
    # for key in skills:
    #   data_df.loc[key] = lst
    # display(data_df)

    #Calculating similarity scores
    sent_embeddings = model.encode(skill_discs)
    idx=0
    for key_embedding in sent_embeddings:
        sim_score_lst = []
        for embedding in sent_embeddings:
          sim_score = 1 - spatial.distance.cosine(key_embedding, embedding)
          sim_score_lst.append(sim_score)
        data_df.loc[skills[idx]] = sim_score_lst
        idx+=1
    return data_df

def filt_top_skills(data_df,n_skills=6):
  #changing data type of dataframe values
  data_df[data_df.columns] = data_df[data_df.columns].astype(float)
  # print(data_df.dtypes)

  sim_skills_dict = {}
  for col in data_df.columns:
    maxValueIndexObj = data_df[col].nlargest(n=5)
    sim_skills_dict[col] = list(maxValueIndexObj.index)[1:]

  return sim_skills_dict
```

```python
skills,skill_discs = load_data('data.json')
data_df = create_df(skills,skill_discs)
sim_skills_dict = filt_top_skills(data_df)
print(sim_skills_dict)

with open('top_skills.json', 'w') as fp1:
    # json.dump(sim_skills_dict, fp1)
    fp1.write(json.dumps(sim_skills_dict, indent=4))
```
